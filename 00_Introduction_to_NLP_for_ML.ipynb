## 1. What is Natural Language Processing? <a id='what-is-nlp'></a>

**Natural Language Processing (NLP)** is a field of Artificial Intelligence that focuses on enabling computers to understand, interpret, and generate human language.

### Key Aspects of NLP:

- **Understanding**: Extracting meaning from text (e.g., sentiment analysis)
- **Processing**: Manipulating and transforming text data (e.g., translation)
- **Generation**: Creating human-readable text (e.g., chatbots)

### Why is NLP Challenging?

Human language is:
- **Ambiguous**: Words can have multiple meanings
- **Context-dependent**: Meaning changes based on context
- **Complex**: Grammar rules, idioms, sarcasm, etc.
- **Ever-evolving**: New words and phrases emerge constantly

### Example of Ambiguity:
```
"I saw a man on a hill with a telescope."
```
- Did I use a telescope to see the man?
- Was the man holding a telescope?
- Was there a telescope on the hill?

## 7. Simple NLP Example <a id='example'></a>

Let's create a simple end-to-end NLP example to demonstrate the basic pipeline.

**Task**: Analyze the sentiment of movie reviews (positive or negative)

````python
# Step 1: Create sample data
# In a real project, you would load this from a file or database

# Sample movie reviews with their sentiment labels
reviews = [
    "This movie was absolutely fantastic! I loved every moment.",
    "Terrible film, waste of time and money.",
    "Amazing storyline and brilliant acting performances.",
    "Boring and predictable, would not recommend.",
    "One of the best movies I have ever seen!",
    "Disappointing and poorly executed."
]

# Labels: 1 = positive, 0 = negative
labels = [1, 0, 1, 0, 1, 0]

# Create a DataFrame for better visualization
df = pd.DataFrame({
    'review': reviews,
    'sentiment': labels
})

# Map numeric labels to text for clarity
df['sentiment_label'] = df['sentiment'].map({1: 'Positive', 0: 'Negative'})

print("Sample Dataset:")
print("=" * 80)
df
````

````python
# Step 2: Text Preprocessing
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import string

def preprocess_text(text):
    """
    Preprocess a text string:
    1. Convert to lowercase
    2. Tokenize into words
    3. Remove punctuation
    4. Remove stopwords (common words like 'the', 'is', 'and')
    
    Args:
        text (str): Input text to preprocess
    
    Returns:
        str: Cleaned and preprocessed text
    """
    # Convert to lowercase
    text = text.lower()
    
    # Tokenize: split text into individual words
    tokens = word_tokenize(text)
    
    # Remove punctuation and stopwords
    stop_words = set(stopwords.words('english'))
    
    # Filter out stopwords and punctuation
    cleaned_tokens = [
        word for word in tokens 
        if word not in stop_words and word not in string.punctuation
    ]
    
    # Join tokens back into a string
    return ' '.join(cleaned_tokens)

# Apply preprocessing to all reviews
df['cleaned_review'] = df['review'].apply(preprocess_text)

print("Before and After Preprocessing:")
print("=" * 80)
for idx, row in df.iterrows():
    print(f"\nOriginal: {row['review']}")
    print(f"Cleaned:  {row['cleaned_review']}")
    print("-" * 80)
````

````python
# Step 3: Feature Extraction - Convert text to numbers
# We'll use CountVectorizer (Bag of Words approach)
from sklearn.feature_extraction.text import CountVectorizer

# Initialize the vectorizer
# This will create a vocabulary of all unique words and count their occurrences
vectorizer = CountVectorizer()

# Fit the vectorizer on our cleaned reviews and transform them to numerical vectors
X = vectorizer.fit_transform(df['cleaned_review'])

# Get the feature names (vocabulary)
feature_names = vectorizer.get_feature_names_out()

print(f"Vocabulary size: {len(feature_names)} unique words\n")
print(f"Vocabulary: {list(feature_names)}\n")

# Convert to array for display
X_array = X.toarray()

# Create a DataFrame to show the feature matrix
feature_df = pd.DataFrame(X_array, columns=feature_names)

print("Feature Matrix (Bag of Words):")
print("Each row is a review, each column is a word, values are word counts")
print("=" * 80)
feature_df
````

````python
# Step 4: Visualize the word frequencies
import matplotlib.pyplot as plt

# Sum the word counts across all reviews
word_counts = X_array.sum(axis=0)

# Create a DataFrame for easier plotting
word_freq_df = pd.DataFrame({
    'word': feature_names,
    'frequency': word_counts
}).sort_values('frequency', ascending=False)

# Plot the top 15 most frequent words
plt.figure(figsize=(12, 6))
plt.bar(word_freq_df.head(15)['word'], word_freq_df.head(15)['frequency'], color='steelblue')
plt.xlabel('Words', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.title('Top 15 Most Frequent Words in Reviews', fontsize=14, fontweight='bold')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.grid(axis='y', alpha=0.3)
plt.show()

print("\nWord Frequency Table:")
print(word_freq_df)
````

````python
# Step 5: Train a simple Machine Learning model
# Note: This is a toy example with very little data
# In practice, you need much more data for good results

from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report

# For this small example, we'll train on all data
# Normally, you would split into train/test sets

# Initialize the Naive Bayes classifier
# This is a simple probabilistic classifier that works well with text
classifier = MultinomialNB()

# Train the model
# X: our numerical features (word counts)
# y: our labels (positive/negative)
y = df['sentiment']
classifier.fit(X, y)

# Make predictions on the same data (just for demonstration)
predictions = classifier.predict(X)

print("Model Training Complete!\n")
print("=" * 80)
print("\nPredictions vs Actual:")
for idx, row in df.iterrows():
    pred_label = 'Positive' if predictions[idx] == 1 else 'Negative'
    actual_label = row['sentiment_label']
    match = "‚úì" if pred_label == actual_label else "‚úó"
    print(f"{match} Review: {row['review'][:50]}...")
    print(f"  Predicted: {pred_label}, Actual: {actual_label}\n")
````

````python
# Step 6: Test with new reviews

# Create some new reviews to test our model
new_reviews = [
    "This is an excellent movie with great performances!",
    "Awful experience, very disappointing.",
    "Absolutely loved it, highly recommended!"
]

print("Testing with New Reviews:")
print("=" * 80)

for review in new_reviews:
    # Preprocess the new review
    cleaned = preprocess_text(review)
    
    # Convert to numerical features using the same vectorizer
    features = vectorizer.transform([cleaned])
    
    # Make prediction
    prediction = classifier.predict(features)[0]
    
    # Get prediction probability
    probability = classifier.predict_proba(features)[0]
    
    sentiment = "Positive" if prediction == 1 else "Negative"
    confidence = max(probability) * 100
    
    print(f"\nReview: {review}")
    print(f"Cleaned: {cleaned}")
    print(f"Prediction: {sentiment} (Confidence: {confidence:.2f}%)")
    print("-" * 80)
````
